```python
# Importando as bibliotecas necessaria
from astroquery.gaia import Gaia
import pandas as pd
```

```python
job = Gaia.launch_job("SELECT * FROM gaiadr3.gaia_source WHERE parallax > 0.5 AND mh_gspphot IS NOT NULL ")  # Filtrar por parallax > 5 seleciona apenas estrelas muito próximas do Sol (dentro de ~200 anos-luz).
data = job.get_results()  # retorna uma tabela no formato astropy.table.Table
```

```python
# Vamos Converter para Pandas DataFrame
df = data.to_pandas()
```

```python
df.columns
```

```python
features_desejadas = ['phot_g_mean_mag', 'bp_rp', 'parallax', 'mh_gspphot', 'spectral_type']

# Verificando quais existem
features_existentes = [f for f in features_desejadas if f in df.columns]
features_ausentes = [f for f in features_desejadas if f not in df.columns]

print("Features existentes:", features_existentes)
print("Features ausentes:", features_ausentes)
```

```python
for col in df.columns:
    if "spectraltype_esphs" in col.lower():  # troca por parte do nome da feature que você procura
        print(col)
```
